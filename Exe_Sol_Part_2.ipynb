**2.** Make a function named mean_of_X compute average of X.

```python
X = np.array([4, 10, 12, 23, -2, -1, 0, 0, 0, -6, 3, -7])
```

**<font color='red'>Sol</font>**:

```python
def mean_of_X(nums):
    
    sum_num = 0
    
    for num in nums:
        
        sum_num = sum_num + num 
    
    avg = sum_num/len(nums)
    
    return avg

print(mean_of_X(X))
```

**3.** 

a ) Define function with the following equation:

$$f(x)=x(x-2) + sin(x)$$

**<font color='red'>Sol</font>**:

```python
import numpy as np

def f(x):
    return x*(x - 2) + np.sin(x)
```

b) Define X with range from -0.5 to 3 and dicritized to 100 equal values. And evaluate Y value, Y=f(X)

**<font color='red'>Sol</font>**:

```python
X = np.linspace(-0.5, 3.0, 100) 

Y = f(X)  # evaluate f on the Xpoints
```

c) Plot Y versus X

**<font color='red'>Sol</font>**:

```python
fig, ax = plt.subplots()
ax.plot(X,Y)
```

d) add the label of $y=x(x-2) + sin(x)$ as title and $x$ and $y$ as labels 

**<font color='red'>Sol</font>**:

```python
fig, ax = plt.subplots()
ax.plot(X,Y)
ax.set_title("Graph of $y = x(x-2)+sin(x)$") #LaTex used for equations
ax.set_xlabel("x")
ax.set_ylabel("y")
```
